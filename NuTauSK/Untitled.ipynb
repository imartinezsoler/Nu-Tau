```python
import numpy as np 
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import scipy.interpolate
import pandas as pd
from matplotlib.pyplot import figure
from scipy import interpolate
import h5py
import time
from time import sleep
from tqdm import tqdm
import math
from scipy.interpolate import interp1d
from scipy.interpolate import CubicSpline
from array import array
from scipy.optimize import fsolve
from scipy.optimize import curve_fit
```

```python
Navg = 6.022e23    #avogadro number 

mHK = 180e9        #HyperK mass (180kton) in grams

Time = 10 * 365 * 24 * 3600
```

```python
Flux = pd.read_csv('./Nu_Tau_prompt_H3a_SIBYLL23C_pr.txt', sep='\s+', engine='python')
```

```python
FluxConv = pd.read_csv('./Atm_Combentional.txt', sep='\s+', engine='python')
```

```python
FluxE = np.zeros(1000)
AFluxE = np.zeros(1000)

Fluxene = np.zeros(1000)

dltc = Flux['czen'][1]-Flux['czen'][0]


for i in range(1000):
    for j in range(100):
        
        ind = i * 100 + j
        
        Fluxene[i] = Flux['ene'][ind]
        FluxE[i] += dltc * Flux['nu'][ind]
        AFluxE[i] += dltc * Flux['anu'][ind]
```

```python
FluxCE = np.zeros(1000)
AFluxCE = np.zeros(1000)
FluxCM = np.zeros(1000)
AFluxCM = np.zeros(1000)
FluxCT = np.zeros(1000)
AFluxCT = np.zeros(1000)

FluxeneC = np.zeros(1000)

dltc = FluxConv['czen'][1]-FluxConv['czen'][0]


for i in range(1000):
    for j in range(100):
        
        ind = i * 100 + j
        
        FluxeneC[i] = FluxConv['ene'][ind]
        FluxCE[i] += dltc * FluxConv['e'][ind]
        AFluxCE[i] += dltc * FluxConv['ebar'][ind]
        FluxCM[i] += dltc * FluxConv['m'][ind]
        AFluxCM[i] += dltc * FluxConv['mbar'][ind]
        #FluxCT[i] += dltc * FluxConv['t'][ind]
        #AFluxCT[i] += dltc * FluxConv['tbar'][ind]
```

```python
plt.yscale('log')
plt.xscale('log')
plt.plot(Fluxene, FluxE)
plt.plot(FluxeneC, FluxCE+AFluxCE)
plt.plot(FluxeneC, FluxCM+AFluxCM)
#plt.plot(FluxeneC, FluxCT+AFluxCT)
```

```python
XsecT = np.zeros(100)

EneX = np.linspace(20, 2000, num=100)

for i in range(1,101):
    xsec = pd.read_csv('./Xsec/Events_Neu'+str(i)+'.root.txt', sep='\s+', engine='python')
    XsecT[i-1] = np.sum(xsec['sigma[cm2]'])

```

```python
plt.xscale('log')
plt.yscale('log')
plt.plot(EneX,XsecT)
plt.xlabel('E (GeV)')
plt.ylabel(r'$\sigma$')
```

```python
xsecTot = interpolate.interp1d(EneX, XsecT)
fluxN = interpolate.interp1d(Fluxene, FluxE)
fluxAN = interpolate.interp1d(Fluxene, AFluxE)

fluxCE = interpolate.interp1d(Fluxene, FluxCE)
AfluxCE = interpolate.interp1d(Fluxene, AFluxCE)
fluxCM = interpolate.interp1d(Fluxene, FluxCM)
AfluxCM = interpolate.interp1d(Fluxene, AFluxCM)
```

```python
np.power(3,2)
```

```python
#Ref gamma = -2.37

def Astrf(ene, gamma):
    flux =  4 * np.pi * (1.44e-18) * (np.power(ene/1e5, gamma))
    return flux
```

```python
fluxAst1 = Astrf(Fluxene, -2)
fluxAst2 = Astrf(Fluxene, -2.37)
fluxAst3 = Astrf(Fluxene, -3.4)
```

```python
plt.xscale('log')
plt.yscale('log')
plt.plot(Fluxene, fluxAst2, label='Astro')
plt.plot(Fluxene, FluxE + AFluxE, label='Prompt')
plt.plot(Fluxene, FluxCE+AFluxCE + FluxCM+AFluxCM, label='Convent')
plt.ylabel(r'$\phi_{\nu_{\tau}}$ ($cm^2$ s GeV)$^{-1}$')
plt.xlabel('E(GeV)')
plt.legend()
```

```python
plt.xscale('log')
plt.yscale('log')
plt.plot(Fluxene, fluxAst2, label=r'Astro $\gamma =-2.37$')
plt.plot(Fluxene, fluxAst1, label=r'Astro $\gamma =-2$')
plt.plot(Fluxene, fluxAst3, label=r'Astro $\gamma =-3.4$')
plt.plot(Fluxene, FluxCE+AFluxCE + FluxCM+AFluxCM, label='Convent')
plt.ylabel(r'$\phi_{\nu_{\tau}}$ ($cm^2$ s GeV)$^{-1}$')
plt.xlabel('E(GeV)')
plt.legend()
```

```python
InteE = np.logspace(np.log10(20), np.log10(1999), 1000)
```

```python
EventsC = np.zeros(len(InteE))
EventsP = np.zeros(len(InteE))
EventsA2 = np.zeros(len(InteE))
EventsA1 = np.zeros(len(InteE))
EventsA3 = np.zeros(len(InteE))
EventsPT = 0
EventsA2T = 0
EventsA1T = 0
EventsA3T = 0
EventsCT = 0

dlte = np.log10(InteE[1]) - np.log10(InteE[0])

m_ln10 = np.log(10)

for i in range(len(InteE)):
    EventsP[i] = xsecTot(InteE[i]) * (fluxN(InteE[i]) + fluxAN(InteE[i])) * InteE[i] * m_ln10 * dlte
    EventsA2[i] = xsecTot(InteE[i]) * Astrf(InteE[i], -2.37) * InteE[i] * m_ln10 * dlte
    EventsA1[i] = xsecTot(InteE[i]) * Astrf(InteE[i], -2) * InteE[i] * m_ln10 * dlte
    EventsA3[i] = xsecTot(InteE[i]) * Astrf(InteE[i], -3.4) * InteE[i] * m_ln10 * dlte
    EventsC[i] = xsecTot(InteE[i]) * (fluxCE(InteE[i]) + AfluxCE(InteE[i]) + fluxCM(InteE[i]) + AfluxCM(InteE[i])) * InteE[i] * m_ln10 * dlte
    EventsPT += EventsP[i]
    EventsA1T += EventsA1[i]
    EventsA2T += EventsA2[i]
    EventsA3T += EventsA3[i]
    EventsCT += EventsC[i]
    
    
EventsP *= (Time * mHK * Navg/16)    
EventsPT *= (Time * mHK * Navg/16)
EventsA1 *= (Time * mHK * Navg/16)
EventsA2 *= (Time * mHK * Navg/16)
EventsA3 *= (Time * mHK * Navg/16)
EventsA1T *= (Time * mHK * Navg/16)
EventsA2T *= (Time * mHK * Navg/16)
EventsA3T *= (Time * mHK * Navg/16)
EventsC *= (Time * mHK * Navg/16)    
EventsCT *= (Time * mHK * Navg/16)
```
